```python
#X write a program to get the square of only even numbers from the range given by the user, involving user defined function + 

def square_even(x,y):
    square_even_list = list()
    for i in range(x,y+1) :
        if i%2==0 :
            square_even_list.append(i)
    
    return square_even_list

square_even(2,10)

```

```python
#X From a deep well, a frog jumps 1m up on first day and goes half of the distance down on the next day. Find the height it achieved
#  on day 'n' by user defined functions. Only two decimal places
 
def height_achieved(n):
    height=0 
    for i in range (n+1) :
        if i%2 != 0 :
            height+=1
        else :
            height = height / 2

    return height 


        

height_achieved(100000)
```

```python
# Write a program to get all the 3-digit nos in which all the digits are unique and 
# the product of the digits will be a two digit number

# All digits are unique 
# product of digits is a two digit number 

def check_number(n):
    list = []
    for i in range (2,-1,-1):  
        step =  n//(10**i)
        last_element = step % 10 
        list.append(last_element)

    unique = True 
    for i in range(len(list)) : 
        for j in range(i+1,len(list)) :
            if list[i] == list[j] :
                unique = False 
                break 

    two_digit = False 
    p=1
    for i in list :
        p = p*i 
    
    if p in range(10,100):
        two_digit = True


    if unique and two_digit :
        print ("pass")

    else : 
        print ("fail")


    


check_number(981)


def check_number_gpt(n):
    # Extract digits
    digits = [int(digit) for digit in str(n)]
    
    # Check if all digits are unique
    unique = len(digits) == len(set(digits))
    
    # Calculate the product of the digits
    product = 1
    for digit in digits:
        product *= digit
    
    # Check if the product is a two-digit number
    two_digit = 10 <= product < 100
    
    # Determine if the number passes both conditions
    if unique and two_digit:
        print("pass")
    else:
        print("fail")

# Test the function
check_number_gpt(456)

    
```
